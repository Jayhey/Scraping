```python
title = []
sub = []
sub_1 = []
sub_2 = []
sub_3 = []
price = []
seller = []
phone = []
url = 'http://www.bobaedream.co.kr/cyber/CyberCar.php?gubun=K'

source_code = requests.get(url)
plain_text = source_code.content
soup = BeautifulSoup(plain_text,'lxml',from_encoding='utf-8')

for i in range(len(soup.findAll('tr'))):
    if soup.findAll('tr')[i].find('td','carinfo') is not None:
        title.append(soup.findAll('tr')[i].find('td','carinfo').find('a','title').text)
        sub.append(soup.findAll('tr')[i].find('td','carinfo').find('a','sub').text)
        sub_1.append(soup.findAll('tr')[i].find('td','carinfo').find('a','sub_01').text.split('ㅣ')[0])
        sub_2.append(soup.findAll('tr')[i].find('td','carinfo').find('a','sub_01').text.split('ㅣ')[1])
        sub_3.append(soup.findAll('tr')[i].find('td','carinfo').find('a','sub_01').text.split('ㅣ')[2])
        seller.append(soup.findAll('tr')[i+1].find('td').text)
        if len(soup.findAll('tr')[i+2].find('td').text.split('-'))==3:
            phone.append(soup.findAll('tr')[i+2].find('td').text)
        else:
            phone.append('없음')
```

```python
import string

alpha = string.ascii_uppercase
alpha
```

```python
'''      ctype = 'A'(고정)

      F_except_1 = 'G'(고정)

      cartype = A~F

      years = A~J(E는 제외)

      age = 1~1 if years = A, B, C, D

      age = 1~2 else if years = H

      age = 1~3 else

      fam_type = A~C

      sex = A~B

      F_except = F~G
      
      params = {'ctype':'A','F_except_1':'G' ,'cartype':'A','years':'A','age':'1','fam_type':'A','sex':'A','F_except':'F'}
'''

combinations = []
ctype = 'A'
F_except_1 = 'G'
cartype = alpha[:6]
years = "".join(alpha[:10].split('E'))
age = list(range(1,4))
fam_type = alpha[:3]
sex = alpha[:2]
F_except = alpha[5:7] # F , G


for car in cartype:
    
    combination = ['','','','','','','','']
    combination[0] = ctype
    combination[1] = F_except_1
    combination[2] = car
    
    for year in years:
        if year == 'A' or year == 'B' or year == 'C' or year == 'D':
            combination[3] = year
            combination[4] = age[0]
            for fam in fam_type:
                combination[5] = fam
                for s in sex:
                    combination[6] = s
                    for F in F_except:
                        combination[7] = F
                        combinations.append(combination.copy())
                        
        elif year == 'H':
            combination[3] = year
            for a in age[:2]:
                combination[4] = a
                for fam in fam_type:
                    combination[5] = fam
                    for s in sex:
                        combination[6] = s
                        for F in F_except:
                            combination[7] = F
                            combinations.append(combination.copy())

        else:
            combination[3] = year
            for a in age:
                combination[4] = a
                for fam in fam_type:
                    combination[5] = fam
                    for s in sex:
                        combination[6] = s
                        for F in F_except:
                            combination[7] = F
                            combinations.append(combination.copy())

len(combinations)
```

```python
for c in combinations:
    params = {'ctype':c[0],'F_except_1':c[1] ,'cartype':c[2],'years':c[3],'age':c[4],'fam_type':c[5],'sex':c[6],'F_except':c[7]}
    
    soup = souping(params)
    table = soup.find('table','etc_table1 group_tb search_date')

    columns = []
    for item in table.find('thead').findAll('th'):
        columns.append(item.text)

    value = defaultdict(list)

    for i,item in enumerate(table.find('tbody').findAll('td')[:36]):
        if i%4 == 0:
            value[columns[0]].append(item.text)
        elif i%4 == 1:
            value[columns[1]].append(item.text)
        elif i%4 == 2:
            value[columns[2]].append(item.text)
        elif i%4 == 3:
            value[columns[3]].append(item.text)
    pd.DataFrame(value).to_csv('C:/Users\Administrator\Desktop\할미새사촌/%s_%s_%s_%s_%d_%s_%s_%s.csv'%tuple((item for item in c)),index=False)
```
