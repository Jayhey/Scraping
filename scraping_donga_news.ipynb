```python
def scrap_contents(data, start, finish):
    with progressbar.ProgressBar(max_value=finish-start) as bar:
        for i, x in enumerate(data.link[start:finish]):
            
            url = x
            source_code = requests.get(url)
            plain_text= source_code.content
            soup = BeautifulSoup(plain_text, 'lxml', from_encoding = "utf-8")
            article_content = soup.find('div', class_ = 'article_txt')
            [s.extract() for s in article_content('script')]
            [s.extract() for s in article_content('div', id = "bestnews_layer")]
            [s.extract() for s in article_content('span')]
            [s.extract() for s in article_content('strong', class_ = 'sub_title')]
            [s.extract() for s in article_content('div', class_ = 'article_keyword')]
            [s.extract() for s in article_content('div', class_ = 'article_relation')]
            donga.content[i+start] += article_content.get_text()
            #time.sleep(1)
            bar.update(i)
            
            
```
